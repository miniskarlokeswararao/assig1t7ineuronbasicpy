```python
print("Write a Python Program to Find LCM?")
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def lcm(a, b):
    return (a * b) // gcd(a, b)

def find_lcm(numbers):
    lcm_result = 1
    for num in numbers:
        lcm_result = lcm(lcm_result, num)
    return lcm_result

# Taking input from the user
n = int(input("Enter the count of numbers: "))
numbers = []
for i in range(n):
    num = int(input("Enter number " + str(i+1) + ": "))
    numbers.append(num)

result = find_lcm(numbers)
print("The LCM of", numbers, "is", result)
```

```python
# Convert the integral part to binary
def decimal_to_binary_octal_hexadecimal(decimal):
    integral_part = int(decimal)
    binary = bin(integral_part)[2:]
# Convert the fractional part to binary
    fractional_part = decimal - integral_part
    fractional_binary = ""
    while fractional_part != 0:
        fractional_part = fractional_part*2
        bit = int(fractional_part)
        fractional_binary += str(bit)
        fractional_part -= bit
    # Convert the integral part to octal
        octal = oct(integral_part)[2:]
    # Convert the integral part to hexadecimal
        hexadecimal = hex(integral_part)[2:]
    return binary, fractional_binary, octal, hexadecimal

# Example usage
decimal_number = 10.75
binary, fractional_binary, octal, hexadecimal = decimal_to_binary_octal_hexadecimal(decimal_number)

print("Binary:", binary)
print("Fractional Binary:", fractional_binary)
print("Octal:", octal)
print("Hexadecimal:", hexadecimal)
```

```python
#3.	Write a Python Program for array rotation?
import numpy as np

def rotate_array(arr, positions):
    # Calculate the number of positions to rotate
    rotations = positions % len(arr)

    # Rotate the array using slicing
    rotated_arr = np.concatenate((arr[rotations:], arr[:rotations]))

    return rotated_arr

# Example usage
arr = np.array([1, 2, 3, 4, 5])
positions = 2

# Rotate the array
rotated = rotate_array(arr, positions)

# Print the rotated array
print("Original array:", arr)
print("Rotated array:", rotated)
```

```python
import numpy as np

def split_and_add(arr, split_index):
    # Split the array into two parts
    first_part = arr[:split_index]
    second_part = arr[split_index:]

    # Add the first part to the end of the second part
    new_arr = np.concatenate((second_part, first_part))

    return new_arr

# Example usage
arr = np.array([1, 2, 3, 4, 5])
split_index = 2

# Split the array and add the first part to the end
new_arr = split_and_add(arr, split_index)

# Print the modified array
print("Original array:", arr)
print("Modified array:", new_arr)
```

```python
def is_monotonic(arr):
    increasing = decreasing = True

    # Check if the array is increasing
    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            increasing = False
            break

    # Check if the array is decreasing
    for i in range(len(arr) - 1):
        if arr[i] < arr[i + 1]:
            decreasing = False
            break

    # Return True if the array is either increasing or decreasing
    return increasing or decreasing

# Example usage
arr1 = [1, 2, 3, 4, 5]
arr2 = [5, 4, 3, 2, 1]
arr3 = [1, 2, 3, 2, 1]

print(is_monotonic(arr1))  # True
print(is_monotonic(arr2))  # True
print(is_monotonic(arr3))  # False
```
